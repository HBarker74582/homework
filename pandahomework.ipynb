```python
import pandas as pd
import numpy as np
```

```python
file = "HeroesOfPymoli_Resources_purchase_data.csv"
hero_df = pd.read_csv(file)
```

```python
totalplayers = hero_df["SN"].nunique()
total_players_df = pd.DataFrame({"Total Players": [totalplayers]})
total_players_df
```

```python
number_unique_items = hero_df["Item ID"].nunique()
average_price = hero_df["Price"].mean()
total_number_purchases = hero_df["Purchase ID"].count()
total_revenue = hero_df["Price"].sum()

Purchasing_Analysis = pd.DataFrame({
        "Number of Unique Items":[number_unique_items], 
        "Average Purchase Price":[round(average_price, 2)],
        "Number of Purchases":[total_number_purchases],
        "Total Revenue":[total_revenue]})
Purchasing_Analysis
```

```python
newdf = hero_df[["SN","Gender"]]
newold = newdf.drop_duplicates()
```

```python
gendergroup = newold.groupby(["Gender"])
total_count = gendergroup.count()
percent = (total_count/totalplayers)*100

merged_df = pd.merge(total_count, round(percent, 2), on="Gender")
merged_df = merged_df.rename(columns={"SN_x":"Total Count", "SN_y":"Percentage of Players"})
merged_df                           
```

```python
newdf = hero_df.groupby(["Gender"])
purchase_count = newdf["Purchase ID"].count()
avgprice = newdf["Price"].mean()
total_price = newdf["Price"].sum()
avg_total_per = total_price/merged_df["Total Count"]
```

```python
total_df = pd.DataFrame({"Purchase Count" : purchase_count,
                         "Average Purchase Price": round(avgprice, 2),
                         "Total Purchase Value": total_price,
                         "Avg Total Purchase per Person": round(avg_total_per, 2)
    
    
})
total_df
```

```python
# Age Demographics

agedf = hero_df[["SN","Age"]]
ageddf = agedf.drop_duplicates()
age_only_df = pd.DataFrame(ageddf)
```

```python
print(age_only_df["Age"].max())
print(age_only_df["Age"].min())
```

```python
bins = [0,9,14,19, 24, 29, 34, 39,40]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
pd.cut(age_only_df["Age"], bins, labels=group_labels).head()
       
```

```python
# Place the data series into a new column inside of the DataFrame
age_only_df["Age Group"] = pd.cut(age_only_df["Age"], bins, labels=group_labels)
ageddf.head(5)
```

```python
# Create a GroupBy object based upon "Age Group"
age_group = age_only_df.groupby("Age Group")

age_group_count = age_group["Age Group"].count()
age_group_percent = age_group_count/totalplayers*100
 
```

```python
summary_age = pd.DataFrame({"Total Count" : age_group_count,
                           "Percentage of Players" : round(age_group_percent, 2),})
summary_age  
```

```python
purchase_age_df = hero_df[["SN","Purchase ID", "Price", "Age"]]
purchase_age_df.head()
```

```python
print(purchase_age_df["Age"].max())
print(purchase_age_df["Age"].min())
```

```python
bins = [0,9,14,19, 24, 29, 34, 39,40]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
pd.cut(purchase_age_df["Age"], bins, labels=group_labels).head()
```

```python
purchase_age_df["Age Group"] = pd.cut(purchase_age_df["Age"], bins, labels=group_labels)
purchase_age_df.head(5)
```

```python
# Create a GroupBy object based upon "Age Group"
new_age_group = purchase_age_df.groupby("Age Group")

new_age_group_count = new_age_group["Purchase ID"].count()
new_avgprice = new_age_group["Price"].mean()
new_total_price = new_age_group["Price"].sum()
new_avg_total_per = new_total_price/age_group_count
```

```python
purchase_age = pd.DataFrame({"Purchase Count" : new_age_group_count,
                         "Average Purchase Price": round(new_avgprice, 2),
                         "Total Purchase Value": new_total_price,
                         "Avg Total Purchase per Person": round(new_avg_total_per, 2)
    
    
})
purchase_age
```

```python
# Top Spenders
top_spend_df = hero_df.groupby(["SN"])
top_count = top_spend_df["Purchase ID"].count()
avg_pur_top = top_spend_df["Price"].mean()
total_price_top = top_spend_df["Price"].sum()


```

```python
top_spenders = pd.DataFrame({"Purchase Count" : top_count,
                         "Average Purchase Price": round(avg_pur_top, 2),
                         "Total Purchase Value": total_price_top})
top_spenders_sorted = top_spenders.sort_values("Total Purchase Value", ascending=False)    
top_spenders_sorted.head(5)
```
